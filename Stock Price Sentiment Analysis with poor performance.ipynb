```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

import re
pd.pandas.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('data.csv', encoding= 'unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.shape
```

```python
df.reset_index(inplace=True)
```

```python
df.head(10)
```

```python
df['combine']=df['Top1']+" "+df['Top2']+" "+df['Top3']+" "+df['Top4']+" "+df['Top5']+" "+df['Top6']+" "+df['Top7']+" "+df['Top8']+" "+df['Top9']+" "+df['Top10']+" "+df['Top11']+" "+df['Top12']+" "+df['Top13']+" "+df['Top14']+" "+df['Top15']+" "+df['Top16']+" "+df['Top17']+" "+df['Top18']+" "+df['Top19']+" "+df['Top20']+" "+df['Top21']+" "+df['Top22']+" "+df['Top23']+" "+df['Top24']+" "+df['Top25']
```

```python
df['combine'][1]
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range (0, len(df)):
    review=re.sub('[^a-zA-Z]', ' ', df['combine'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500, ngram_range=(1,3))
x=cv.fit_transform(corpus).toarray()
```

```python
y=df.Label
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

MULTINOMIAL NAIVE BAYES

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('accuracy_score: ', accuracy_score(y_test, y_pred))
```

PASSIVE AGGRESSIVE CLASSIFIER

```python
from sklearn.linear_model import PassiveAggressiveClassifier
clf=PassiveAggressiveClassifier()
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('accuracy_score: ', accuracy_score(y_test, y_pred))
```
